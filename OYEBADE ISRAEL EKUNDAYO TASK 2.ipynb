## Menu-Driven Calculator Program

```python
# Campus Calculator Program
# This program performs basic arithmetic operations
# and keeps running until the user chooses to exit.

# --------- Function Definitions ---------

def add(a, b):
    """Returns the sum of two numbers"""
    return a + b


def subtract(a, b):
    """Returns the difference between two numbers"""
    return a - b


def multiply(a, b):
    """Returns the product of two numbers"""
    return a * b


def divide(a, b):
    """Returns the division of two numbers safely"""
    if b == 0:
        return "Error: Division by zero is not allowed."
    return a / b


# --------- Main Program Loop ---------

while True:
    print("\n=== Campus Calculator ===")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    choice = input("Choose an operation (1–5): ")

    # Exit condition
    if choice == "5":
        print("Thank you for using the Campus Calculator. Goodbye!")
        break

    # Check for valid choice
    if choice in ["1", "2", "3", "4"]:
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choice == "1":
                print("Result:", add(num1, num2))
            elif choice == "2":
                print("Result:", subtract(num1, num2))
            elif choice == "3":
                print("Result:", multiply(num1, num2))
            elif choice == "4":
                print("Result:", divide(num1, num2))

        except ValueError:
            print("Invalid input! Please enter numeric values only.")

    else:
        print("Invalid choice! Please select a number between 1 and 5.")
```

## Class Result Assistant

```python
"""
Class Result Assistant
A program to help teachers store student scores, calculate averages, and assign grades.
"""

def calculate_average(scores):
    """
    Calculate the average of a list of scores.
    
    Args:
        scores (list): List of numerical scores
    
    Returns:
        float: Average score rounded to 2 decimal places
    """
    if not scores:
        return 0.0
    return round(sum(scores) / len(scores), 2)


def assign_grade(average):
    """
    Assign a letter grade based on the average score.
    
    Args:
        average (float): Student's average score
    
    Returns:
        str: Letter grade (A, B, C, D, or F)
    """
    if average >= 70:
        return 'A'
    elif average >= 60:
        return 'B'
    elif average >= 50:
        return 'C'
    elif average >= 40:
        return 'D'
    else:
        return 'F'


def display_results(students_data):
    """
    Display student results in a neat, formatted table.
    
    Args:
        students_data (dict): Dictionary containing student names and their data
    """
    print("\n" + "="*70)
    print("STUDENT RESULTS SUMMARY".center(70))
    print("="*70)
    print(f"{'Name':<20} {'Scores':<25} {'Average':<10} {'Grade':<5}")
    print("-"*70)
    
    # Loop through each student and display their results
    for name, data in students_data.items():
        scores_str = str(data['scores'])
        print(f"{name:<20} {scores_str:<25} {data['average']:<10} {data['grade']:<5}")
    
    print("="*70 + "\n")


def add_student(students_data):
    """
    Add a new student with their test scores.
    
    Args:
        students_data (dict): Dictionary to store student information
    """
    name = input("Enter student name: ").strip()
    
    # Validate that name is not empty
    if not name:
        print("Error: Name cannot be empty.")
        return
    
    # Get test scores from the teacher
    scores = []
    num_tests = int(input(f"How many test scores for {name}? "))
    
    for i in range(num_tests):
        while True:
            try:
                score = float(input(f"  Enter score {i+1}: "))
                if 0 <= score <= 100:
                    scores.append(score)
                    break
                else:
                    print("  Please enter a score between 0 and 100.")
            except ValueError:
                print("  Invalid input. Please enter a number.")
    
    # Calculate average and assign grade
    average = calculate_average(scores)
    grade = assign_grade(average)
    
    # Store student data in dictionary
    students_data[name] = {
        'scores': scores,
        'average': average,
        'grade': grade
    }
    
    print(f"\n✓ {name} added successfully with average {average} (Grade: {grade})\n")


def main():
    """
    Main function to run the Class Result Assistant program.
    """
    # Dictionary to store all student data
    students_data = {}
    
    print("\n" + "="*70)
    print("CLASS RESULT ASSISTANT".center(70))
    print("="*70)
    print("Welcome! This program helps you manage student scores and grades.\n")
    
    # Main program loop
    while True:
        print("\nOptions:")
        print("  1. Add a student")
        print("  2. View all results")
        print("  3. Exit")
        
        choice = input("\nEnter your choice (1-3): ").strip()
        
        if choice == '1':
            add_student(students_data)
        
        elif choice == '2':
            if students_data:
                display_results(students_data)
            else:
                print("\nNo students added yet.\n")
        
        elif choice == '3':
            print("\nThank you for using Class Result Assistant!")
            print("Goodbye!\n")
            break
        
        else:
            print("\nInvalid choice. Please enter 1, 2, or 3.\n")


# Run the program
if __name__ == "__main__":
    main()
```
